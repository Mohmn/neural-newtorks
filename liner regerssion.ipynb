```python
from matplotlib import pyplot as plt
import math
from random import *
m=1
b=0

def lr(x,y):
    xmean=0
    ymean=0
    for i in range(len(x)):
        xmean+=x[i]
        ymean+=y[i]
    xmean = xmean/len(x)
    ymean = ymean/len(y)
    num,dem=0,0
    for i in range(len(x)):
        
        num += (x[i]-xmean)*(y[i] - ymean)
        dem += ((x[i]-xmean)**2)
    m = num/dem
    b= ymean - (xmean * m)
    #print(m,b)
    #print(x,y)
    
    for i in range(len(y)):
        y[i] = m*x[i] + b
    
    plt.plot(x,y,color='g',linewidth=1)
    plt.figure(figsize=(70,70))
    #print(x,y)
    
x=[randint(0,50) for i in range(1000)]
y=[randint(-1000,1000) for i in range(1000)]

plt.plot(x,y,'ro',ms=5)

lr(x,y)

plt.figure(figsize=(70,70))
plt.show()
```
