```python
from matplotlib import pyplot as plp
import math,csv
from random import *

def error(x,y,m,b):
    error = 0
    for i in range(len(x)):
        error += (y[i] - (m*x[i] + b))**2
    print(error,"err")
def gradient_descent(x,y,m,b):
    statring_m=0
    statring_b=0
    n=len(x)
    lr = 0.01
    for i in range(len(x)): 
        statring_m+= (y[i] - (m*x[i] + b))*(x[i])
        statring_b += (y[i] - (m*x[i] + b))
    
    statring_m *=  (-1/n)
    statring_b*=(-1/n)
    new_m = m -(lr*statring_m)
    new_b = b - (lr*statring_b)
    return new_m ,new_b
def drawline(x,y,color,m,b):
    print(m,b)
    x1,y1 = [],[]
    x1.append(0)
    y1.append(m*0+b)
    c=max(x)
    x1.append(c)
    y1.append(m*c+b)
    #print(y1)
    plp.plot(x1,y1,color = color)

        
def main():
    c = open("ex1data1.txt")
    f = csv.reader(c)
    x,y=[],[]
    for a,b in f:
        x.append(float(a))
        y.append(float(b))
    
    #print(x,y)
    
    plp.figure(figsize=(10,10))
    plp.plot(x,y,"ro",ms='1.5')
    
    
    m,b=0,0
    error(x,y,m,b)
    for i in range(100000):
        m,b = gradient_descent(x,y,m,b)
        #error(x,y,m,b)
    
    error(x,y,m,b)
    drawline(x,y,"g",m,b)
    
main()
```
